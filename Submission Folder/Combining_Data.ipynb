### Importing CSVs 
1. Lorenz group
2. PLAY NEWS group
3. Medusa group
4. 8Base group
5. Lockbit group


```python
Lorenz = pd.read_csv("Lorenz.csv")
PlayNews = pd.read_csv("PlayNews.csv")
EightBase = pd.read_csv("8Base.csv")

with open('Medusa.csv', 'rb') as f:
    result = chardet.detect(f.read())
Medusa = pd.read_csv('Medusa.csv', encoding=result['encoding'])
```

```python
Lorenz = Lorenz[['Company_Name', 'Date_of_leak', 'Company_Address', 'Company_Country', 'Industry_Type', 'Personal Identifiable Information (PII)','Financial Data','Healthcare Information','Employee and HR Data','Intellectual Property and Trade Secrets','Customer and Client Data','Legal Documents','Government and Defense Information','Educational Data','Credentials and Access Information','Databases','Website Data','Backup Data','Video and Surveillance Footage','Social Media Data','Utility and Infrastructure Data','Sensitive Communication']]
llist = ["Lorenz"]*Lorenz.shape[0]
Lorenz['Ransomware_Group'] = llist
```

```python
EightBase =EightBase[['Company_Name', 'Date_of_leak', 'Company_Address', 'Company_Country', 'Industry_Type', 'Personal Identifiable Information (PII)','Financial Data','Healthcare Information','Employee and HR Data','Intellectual Property and Trade Secrets','Customer and Client Data','Legal Documents','Government and Defense Information','Educational Data','Credentials and Access Information','Databases','Website Data','Backup Data','Video and Surveillance Footage','Social Media Data','Utility and Infrastructure Data','Sensitive Communication']]
llist = ["8Base"]*EightBase.shape[0]
EightBase['Ransomware_Group'] = llist
```

```python
PlayNews =PlayNews[['Company_Name', 'Date_of_leak', 'Company_Address', 'Company_Country', 'Industry_Type', 'Personal Identifiable Information (PII)','Financial Data','Healthcare Information','Employee and HR Data','Intellectual Property and Trade Secrets','Customer and Client Data','Legal Documents','Government and Defense Information','Educational Data','Credentials and Access Information','Databases','Website Data','Backup Data','Video and Surveillance Footage','Social Media Data','Utility and Infrastructure Data','Sensitive Communication']]
llist = ["PlayNews"]*PlayNews.shape[0]
PlayNews['Ransomware_Group'] = llist
```

```python
Medusa =Medusa[['Company_Name', 'Date_of_leak', 'Company_Address', 'Company_Country', 'Industry_Type', 'Personal Identifiable Information (PII)','Financial Data','Healthcare Information','Employee and HR Data','Intellectual Property and Trade Secrets','Customer and Client Data','Legal Documents','Government and Defense Information','Educational Data','Credentials and Access Information','Databases','Website Data','Backup Data','Video and Surveillance Footage','Social Media Data','Utility and Infrastructure Data','Sensitive Communication']]
llist = ["Medusa"]*Medusa.shape[0]
Medusa['Ransomware_Group'] = llist
```

### Concatenating all the dataframes

```python
Final_DF = pd.concat([Lorenz, EightBase, PlayNews, Medusa], axis=0)
```

### Analyzing basic graphs (before tableau)

#### Ransomware Group Activity

```python

fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Ransomware_Group.value_counts().index, y=Final_DF.Ransomware_Group.value_counts())
```

#### Industry Type attacked

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Industry_Type.value_counts().index, y=Final_DF.Industry_Type.value_counts())
```

#### Countries Attacked

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Company_Country.value_counts().index, y=Final_DF.Company_Country.value_counts())
```

```python
Final_DF.to_csv("Combined.csv")
```
