```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

### Importing CSVs 
1. Lorenz group
2. PLAY NEWS group
3. Medusa group
4. 8Base group
5. Lockbit group


```python
Lorenz = pd.read_csv("Lorenz.csv")
PlayNews = pd.read_csv("PlayNews.csv")
```

```python
Lorenz = Lorenz[['Company_Name', 'Date_of_leak', 'Data_Type', 'Company_Address', 'Company_Country', 'Industry_Type']]
llist = ["Lorenz"]*Lorenz.shape[0]
Lorenz['Ransomware_Group'] = llist
```

```python
PlayNews = PlayNews[['Company_Name', 'Date_of_leak', 'Data_Type', 'Company_Address', 'Company_Country', 'Industry_Type']]
llist = ["PlayNews"]*PlayNews.shape[0]
PlayNews['Ransomware_Group'] = llist
```

### Concatenating all the dataframes

```python
Final_DF = pd.concat([Lorenz, PlayNews], axis=0)
```

### Analyzing basic graphs (before tableau)

#### Ransomware Group Activity

```python

fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Ransomware_Group.value_counts().index, y=Final_DF.Ransomware_Group.value_counts())
```

#### Industry Type attacked

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Industry_Type.value_counts().index, y=Final_DF.Industry_Type.value_counts())
```

#### Countries Attacked

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
sb.barplot(x=Final_DF.Company_Country.value_counts().index, y=Final_DF.Company_Country.value_counts())
```
